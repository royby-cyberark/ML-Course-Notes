# Data preparing 

delete all the diamonds with more than 10 percent difference between x and y

```python
df.drop(df.query("x/y < 0.9 or x/y > 1.1").index , inplace=True)
```

delete all the diamonds with z>x

```python
df.drop(df.query("z>x").index , inplace=True)
```

Create a new column 'xy' with the mean value of x and y

```python
df['xy']=(df.x + df.y)/2
```

```python
df.head()
```

delete columns x,y

```python
df.drop(['x','y'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.groupby(['color']).mean()
```

Convert the color feature to int value

```python
df.color.replace({'E':1 , 'D':2 , 'F':3 , 'G':4 , 'H':5 , 'I':6 , 'J':7},inplace=True)
```

```python
df.head()
```

delete the columns cut and clarity 

```python
df.drop(['cut','clarity'],axis=1,inplace=True)
```
